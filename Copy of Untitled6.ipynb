```
path =  '/content/gdrive/My Drive/Colab Notebooks/twitter_data'
(x_train, y_train), (x_test, y_test), preproc = text.texts_from_folder(path, 
                                                                       maxlen=25, 
                                                                       preprocess_mode='bert',
                                                                       train_test_names=['train', 
                                                                                         'test'],
                                                                       classes=['1', '2' , '3'])
print('done')
```

```
model = text.text_classifier('bert', (x_train, y_train), preproc=preproc)
learner = ktrain.get_learner(model,train_data=(x_train, y_train), val_data=(x_test, y_test), batch_size=64)
```

```
learner.fit_onecycle(2e-5, 1)
```
